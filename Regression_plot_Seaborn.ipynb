```python
import seaborn as sns
import matplotlib.pyplot as plt
```


**lmplot()**

regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it.

Scatterplot ke andar calculation ke baad ek bestfit line banti hai and use hi hm regression plo bolte hai.

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips)# jisko predict karna hai use hm y per rakhenge
#jisse predict karna hai use x per
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm')#hue se male and female ko a;ah lar diye
```

```python
sns.lmplot(x = 'total_bill',y = 'tip',data = tips,hue = 'sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':100})#o for oval and v is traingle
```
