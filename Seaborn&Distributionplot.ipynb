```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

We use distribution plots to visualize the distribution of quantitative data.

Let's discuss some plots that allow us to visualize the distribution of a data set. These plots are:

distplot/histplot

jointplot

pairplot

rugplot

```python
df = sns.load_dataset('tips')#hm bina dale ise load kar diye kyuki seaborn pass pahle se hi ye data hota hai.
df
```

```python
df['size'].unique()#so this is categorial or qualitative
#and distribution plot is only for quantative
```

```python

sns.histplot(df['total_bill'],bins = 2)
```

```python

sns.histplot(df['total_bill'],bins = 20)
```

```python

sns.histplot(df['total_bill'],bins = 20,kde = True)#kernal density estimator
```

```python
plt.subplot(1,2,1)#ek canvas per do banane ke liye
sns.histplot(df['total_bill'],bins = 20,kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 20,kde = True)
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'scatter')
# scatterplot me sare rows ka dot banta hai and isi ke help se bestfit line banta hai
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill', y = 'tip', data = df,kind = 'reg')
#agar data ka distribution check karna hai tu histogram banao
#agar do quantaative ka distribution check karna hai tu jointplot banao
```

```python
#pairplot
sns.pairplot(df)#sare numericals column ka comparision hua hai.
```

```python

sns.pairplot(df,hue='sex')#blue male and oranage feamlae
#hue differentiate karne ke liye bas
```

```python
#palette colour ke liye
sns.pairplot(df,hue='size',palette='rainbow')
```

```python
sns.rugplot(df['tip'])#jha jha hai wha wha ek line dikha deta hai
```
